Next we have to open the data that we want to analyze. As described above the data is stored in a _*.root_ file. This is a root file containing tracks and calorimeter clusters

```python
#f = ROOT.TFile.Open("Tracks_Clusters.root") 
f = ROOT.TFile.Open("https://csandova.web.cern.ch/csandova/HEP-Ex-Course/Code/TracksClusters/Tracks_Clusters.root") 
```

After the data is opened we create a canvas on which we can draw a histogram. If we do not have a canvas we cannot see our histogram at the end. Its name is _Canvas_ and its header is _a first way to plot a variable_. The two following arguments define the width and the height of the canvas.

```python
canvas = ROOT.TCanvas("Canvas","a first way to plot a variable",800,600)
```

The next step is to define a tree named _tree_ to get the data out of the _*.root_ file. The tree in this root file is called "JetRecoTree"

```python
tree = f.Get("JetRecoTree")
```

Print the number of entries. This is the number of events saved.

```python
tree.Print()
tree.GetEntries()
```

Now we define a histogram that will later be placed on this canvas. Its name is NPV and the header of the histogram is Example plot: Number of primary vertices. The three following arguments indicate that this histogram contains 50 so called bins which have a range from 1 to 50.

```python
hist_npv = ROOT.TH1F("NPV","Example plot: Number of primary vertices; NPV ; Events ",50,1,50)
#hist_npv_mu = ROOT.TH2F("NPV","Example plot: Number of primary vertices; NPV ; Events ",50,1,50,)
```

The following lines are a loop that goes over the data (events) that is stored in the tree and fills the histogram with the event variable NPV. In this first notebook we don't do any cuts to keep it simple. Accordingly the loop fills the histogram for each event stored in the tree. After the program has looped over all the data it prints the word __Done!__.

```python
for event in tree:
    hist_npv.Fill(tree.NPV)
    
print("Done!")
```

```python
hist_npv.SetFillColor(5)
hist_npv.Draw()
```

```python
canvas.Draw()
```

Now we normalise the histogram to the number of entries

```python
scale = hist_npv.Integral()
hist_npv.Scale(1/scale)
hist_npv.SetFillColor(5)
#hist_npv.SetMarkerStyle(1)
```

```python
hist_npv.Draw()
canvas.Draw("hist_npv")
```

**1.** Number of average interactions per bunch-crossing (mu_average):

First, it is necessary to define a new histogram:

```python
hist_muaverage=ROOT.TH1F("mu_average", "Histogram of average interactions per bunch-crossing; mu_average; Events",50,1,100)
```

**2**. A 2D histogram which includes the NPV and mu_average histograms.

Some branches are not event variables, but objects with some kinematical infomation. They are stored in vectors. For example tracks and clusters. We will create two histograms to store the number of tracks and clusters in each event. 

```python
hist_nTrks = ROOT.TH1F("NTracks","Example plot: Number of tracks; NTracks ; Events ",50,0,2000)
hist_nCltrs = ROOT.TH1F("NClusters","Example plot: Number of clusters; NClusters ; Events ",50,0,1500)
```

```python
for event in tree:
    hist_nTrks.Fill(tree.Tracks_pt.size())
    hist_nCltrs.Fill(tree.Clusters_pt.size())
    
print("Done!")
```

```python
hist_nTrks.SetFillColor(4)
hist_nTrks.Draw()
canvas.Draw()
```

$p_T$: Es el momentum transversal al eje z (direccin del haz), i.e., est en el plano xy.

$\eta$: Es la llamada *pseudorapidez*, definida como:

\begin{equation}
\eta = -ln|\tan{\left(\frac{\theta}{2}\right)}|
\end{equation}

siendo $\theta$ el angulo polar, por lo que se mide con respecto al eje z y cuantifica la desviacin de las partculas con respecto a la direccin del haz inicial. En teora, ira desde $-\infty$ hasta $\infty$, pero en la prctica, el rango ATLAS es ms o menos de $|\eta|<5$. 


$\phi$: Es el ngulo azimutal, i.e., el medido en el plano xy.

$m$: Es la masa de la partcula que se detect.

$v_{tx}$: Es el vrtice donde se origin la partcula que se detect.

**6**. 

The PDG (Particle Data Group) ID are codes used in the "PDG particle numbering scheme" which is a way to label particles and objects in particle physics, assigning a unique code number for an specific particle, hadron, etc... And they are widely used in Monte Carlo Event Generators to label the particles which are simulated.

From the bottom line concerning to "Particles_pdgID", it is clear that the code number is **22**, which means the simulated particle corresponds to a photon. This makes sense with the above results in particle_m and clusters_m, where the mass is null. However, in track_m the mass is not zero, but it could be background noisy, since we know that in the tracker is not possible to detect a photon.

In conclusion, the data points out to be a photon.

Information taken from: https://particle.wiki/wiki/PDG_particle_numbering_scheme
